Model 2: albert-base-v2
https://huggingface.co/albert-base-v2


Pretrained model on English language using a masked language modeling (MLM) objective. It was introduced in this paper and first released in this repository. This model, as all ALBERT models, is uncased: it does not make a difference between english and English.

```python
unmasker = pipeline('fill-mask', model='albert-base-v2')
```

```python
list = ["MWEs are a pain in the neck for NLP.", "better safe than sorry.", "read between the lines.", "play your cards right.", "it's an uphill battle.", "you can't judge a book by its cover.", "ignorance is bliss.", "the grass is always greener on the other side."]

print(list[0].split())

def mask_words(string):
    words = string.split()
    masked_strings = []

    for i in range(len(words)-1):
        masked_words = words.copy()
        masked_words[i] = "[MASK]"
        masked_string = " ".join(masked_words)
        masked_strings.append(masked_string)

    masked_strings.append(string.replace(words[len(words)-1], '[MASK].'))
    masked_strings.append(string.replace('.', '[MASK]'))

    return masked_strings

res = []

for string in list:
  res.extend(mask_words(string))

res
```

```python
table_results = []
for string in res:
  table_results.extend(unmasker(string))

table_results
```
